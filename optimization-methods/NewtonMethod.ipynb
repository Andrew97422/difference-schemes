```python
import math
import numpy as np

round = 4
def format_print(x):
    return "{0:.4f}".format(x)


def g(x):
    return x[0] ** 4 + x[1] ** 4 + np.sqrt(2 + x[0] ** 2 + x[1] ** 2) - 2 * x[0] + 3 * x[1]


def diffg(x):
    res1 = 4 * x[0] ** 3 + x[0] / (np.sqrt(2 + x[0] ** 2 + x[1] ** 2)) - 2
    res2 = 4 * x[1] ** 3 + x[1] / (np.sqrt(2 + x[0] ** 2 + x[1] ** 2)) + 3
    res = [res1, res2]
    return res


def diff2g(x):
    res1 = 12 * x[0] ** 2 + (2 + x[0] ** 2 + x[1] ** 2) ** (- 0.5) - x[0] ** 2 * (2 + x[0] ** 2 + x[1] ** 2) ** (- 1.5)
    res2 = - x[0] * x[1] * (2 + x[0] ** 2 + x[1] ** 2) ** (- 1.5)
    res3 = - x[0] * x[1] * (2 + x[0] ** 2 + x[1] ** 2) ** (- 1.5)
    res4 = 12 * x[1] ** 2 + (2 + x[1] ** 2 + x[0] ** 2) ** (- 0.5) - x[1] ** 2 * (2 + x[1] ** 2 + x[0] ** 2) ** (- 1.5)
    res = [[res1, res2], [res3, res4]]
    return res


def det2(matrix):
    return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]


def find_minor(matrix, i, j):
    tmp = [row for k, row in enumerate(matrix) if k != i]
    tmp = [col for k, col in enumerate(zip(*tmp)) if k != j]
    return tmp


def determinant(matrix):
    size = len(matrix)
    if size == 2:
        return det2(matrix)
    return sum((-1) ** j * matrix[0][j] * determinant(find_minor(matrix, 0, j))
               for j in range(size))


def transpose(m):
     return map(list, zip(*m))


def inverse_matrix(m):
    det = determinant(m)
    if det == 0:
        raise ValueError("Матрица вырожденная!")
    if len(m) == 2:
        return [[m[1][1] / det, -1 * m[0][1] / det],
                [-1 * m[1][0] / det, m[0][0] / det]]

    columns = []
    for r in range(len(m)):
        rows = []
        for c in range(len(m)):
            minor = find_minor(m,r,c)
            rows.append(((-1) ** (r + c)) * det(minor))
        columns.append(rows)
    columns = transpose(columns)
    for r in range(len(columns)):
        for c in range(len(columns)):
            columns[r][c] = columns[r][c]/det
    return columns


eps = 10 ** (-round)
xk = np.array([2, 2])
gk = diffg(xk)
dgk = diff2g(xk)
normg = np.sqrt((gk[0] ** 2) + (gk[1] ** 2))
xmin = 0
i = 0
while normg >= eps:
    mat = inverse_matrix(dgk)
    xk = [xk[0] - (mat[0][0] * gk[0] + mat[0][1] * gk[1]), xk[1] - (mat[1][0] * gk[0] + mat[1][1] * gk[1])]
    i += 1
    gk = diffg(xk)
    normg = np.sqrt((gk[0] ** 2) + (gk[1] ** 2))
    print(i, '\tx = [' + format_print(xk[0]) + ', ' + format_print(xk[1]) + ']\t f(x) = [' + format_print(gk[0]) +  ', ' + format_print(gk[1]) + ']\t'"||f'(x)|| = " +
          format_print(normg))
print('xmin = [' + format_print(xk[0]) + ', ' + format_print(xk[1]) + ']\tf(xmin) = [' + format_print(gk[0]) + ', ' + format_print(gk[1]) + ']\t'"||f'(xmin)|| = " +
      format_print(normg))
```
