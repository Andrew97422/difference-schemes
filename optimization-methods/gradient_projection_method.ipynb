```python
import math
import numpy as np

round = 6


def g(x):
    res = np.log(x[0] ** 2 + x[1] ** 2 - 8 * x[0] - 6 * x[1] + 26) - x[0] - x[1]
    return res


def diffg(x):
    res1 = (2 * x[0] - 8) / (x[0] ** 2 + x[1] ** 2 - 8 * x[0] - 6 * x[1] + 26) - 1
    res2 = (2 * x[1] - 6) / (x[0] ** 2 + x[1] ** 2 - 8 * x[0] - 6 * x[1] + 26) - 1
    res = np.array([res1, res2])
    return res


def format_print(x):
    s = "{0:." + str(round) + "f}"
    return s.format(x)


def norma(x):
    return np.sqrt(x[0] ** 2 + x[1] ** 2)


def scalarProduct(a, b):
    return a[0] * b[0] + a[1] * b[1]



def proj(x):
    projection = []

    if x[0] >= 0 and x[0] <= 3 and x[1] >= 0 and x[1] <= 2 and x[0] + x[1] <= 4:
        return x

    if x[0] < 0:
        projection.append(0)
        if x[1] < 0:
            projection.append(0)
        elif x[1] > 2:
            projection.append(2)
        else:
            projection.append(x[1])

    if x[0] >= 0 and x[0] <= 3 and x[1] > 2 and x[0] + x[1] < 4:
        projection.append(x[0])
        projection.append(2)

    if x[0] >= 0 and x[0] <= 3 and x[1] < 0 and x[0] + x[1] < 4:
        projection.append(x[0])
        projection.append(0)

    if x[1] >= 0 and x[1] <= 2 and x[0] > 3 and x[0] + x[1] < 4:
        projection.append(3)
        projection.append(x[1])

    if x[0] > 3 and x[1] < 0 and x[0] + x[1] < 4:
        projection.append(3)
        projection.append(0)

    if x[0] + x[1] > 4:
        pr = [x[0] + (4 - x[0] - x[1]) / 2, x[1] + (4 - x[0] - x[1]) / 2]
        if pr[0] < 0:
            projection.append(0)
        elif pr[0] > 3:
            projection.append(3)
        else:
            projection.append(pr[0])
        if pr[1] > 2:
            projection.append(2)
        elif pr[1] < 0:
            projection.append(0)
        else:
            projection.append(pr[1])

    return projection

def z(n, gradn, s):
    x1 = np.array([n[0] - s * gradn[0], n[1] - s * gradn[1]])
    k = proj(x1)
    return k


def findS(n, gradn, sig, a):
    s = 3
    k = z(n, gradn, s)
    nk = math.sqrt((n[0] - k[0]) ** 2 + (n[1] - k[1]) ** 2)

    while g(n) - g(k) < sig * (nk ** 2):
        s = s * a
        k = z(n, gradn, s)
        nk = (n[0] - k[0]) ** 2 + (n[1] - k[1]) ** 2

    return s


def gradMethod(n, sig, a, e):
    i = 0
    while True:
        gradn = diffg(n)
        s = findS(n, gradn, sig, a)
        k = z(n, gradn, s)
        nk = (n[0] - k[0]) ** 2 + (n[1] - k[1]) ** 2
        norm = norma(gradn)
        print(str(i + 1) + ')\tz = [' + format_print(k[0]) + ', ' + format_print(k[1]) + ']\ts = ' + format_print(s) + "\tf'x = [" +format_print(gradn[0]) + ', ' +
              format_print(gradn[1]) + ']\t|x - z| = ' + format_print(nk))
        if nk > e and norm > e:
            n = k
            i += 1
        else:
            break
    return n
# метод проекции градиента
xk = [-2, 3]
alfa = 0.7
sigma = 0.2
eps = 10 ** (-round)
xm = gradMethod(xk, sigma, alfa, eps)
print('xmin = [' + format_print(xm[0]) + ', ' + format_print(xm[1]) + ']')
```
