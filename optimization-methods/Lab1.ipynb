```
import math

def f(x):
  return math.sqrt(1 + x ** 2) + math.e ** (-2 * x)

def format_print(x):
  return "{0:.6f}".format(x)

a = 0
b = 1
delta = (3 - math.sqrt(5)) * (b - a) / 2
eps = 0.000001
x = a + delta
y = b - delta
xmin = (a + b) / 2
i = 1

while b - a >= 2 * eps:
  print(i, '\t', format_print(a), format_print(b),
        format_print(x), format_print(y),
        format_print(f(x)), format_print(f(y)))
  xmin = (a + b) / 2
  if f(x) < f(y):
    b = y
    y = x
    x = b + a - y
  else:
    a = x
    x = y
    y = b + a - x
  i += 1

print(format_print(xmin))
```
