```python
from math import *
import matplotlib.pyplot as plt
import numpy as np

## Дифференциальное решение
def diff_equation(x):
  return np.sin(x) * np.log(((np.cos(x) - 1) / (np.cos(x) + 1)) ** 2) + (4 - np.log(1)) * np.sin(x)

# Указываем размер сетки
N = 10

# Константы
round = 4
h = 1 / N
A = 1
B = 0
C = 1
a = A / (h ** 2) - B / (2 * h)
b = C - 2 * A / (h ** 2)
c = A / (h ** 2) + B / (2 * h)
round_num = 4

# Иксы
x = []
for i in range(N):
  x.append(np.round(np.pi / 2 + i * h, round))
print(x)
# f n-e
def fn(x):
  return 4 / (np.tan(x))

# list игреков
y_res = [4]
y_diff = [np.round(diff_equation(x[0]), round)]

# Функция для создания игреков
def yn(i):
  if i == 1:  return y_res[0] + 4 * h
  else: return 1 / c * (fn(x[i]) - a * y_res[i - 2] - b * y_res[i - 1])

# Заполнение массива игреков
for i in range(1, len(x)):
  y_res.append(np.round(yn(i), round))
  y_diff.append(np.round(diff_equation(x[i]), round))

#print(y_res)
# Создание массива в нормальном виде для numpy
y_num = np.array(y_res)
y_dif = np.array(y_diff)

print(y_num.size)
print(y_num)
print(y_dif.size)
print(y_dif)
plt.plot(x, y_num, 'c', color='red')
plt.plot(x, y_dif, 'c', color='green')
plt.show()
```
